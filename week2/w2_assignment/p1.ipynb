1)Normalize Customer Age & Salary for Credit Score Prediction
Topics: Normalization (Min-Max), Data Scaling
Problem:
A bank collected customer data containing:
age, monthly_salary, credit_score

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import pandas as pd
df = pd.read_csv("Sample datasets/1.csv")
```

```python
scalar = MinMaxScaler()
df['age'] = scalar.fit_transform(df[['age']])
df['monthly_salary'] = scalar.fit_transform(df[['monthly_salary']])
df
```

3)Binarize Customer Spending for Marketing Segmentation
Topics: Binarization

Problem:
An online store tracks monthly_spending of customers.
Using a binarizer threshold of ₹5000, create:

1 → High spender

0 → Low spender

Use binarization to build this feature.

```python
from sklearn.preprocessing import Binarizer
import pandas as pd
df = pd.read_csv("Sample datasets/3.csv")

binarizer = Binarizer(threshold=5000)
df["monthly_spending"] = binarizer.fit_transform(df[['monthly_spending']])
df

```

4)Predict Student Scores Using Linear Regression
Topics: Linear Regression

Problem:
Given data:
hours_studied → exam_score

Train a Linear Regression model to predict a student’s expected score based on hours studied.


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

l_reg = LinearRegression()
import pandas as pd
df = pd.read_csv("Sample datasets/4.csv")
x = df[['hours_studied']]
y = df['exam_score']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)


l_reg.fit(x_train, y_train)
l_reg.score(x_train, y_train)
import numpy as np
new_predictions = np.array([[9]])
l_reg.predict(new_predictions)
```
