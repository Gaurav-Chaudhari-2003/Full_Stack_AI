```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df = data
df["yt"] = (df["yt"] > 2).astype(int)
df.result = le.fit_transform(df.result)
df
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
```

```python
x = df[['yt','guide']]
y = df['result']
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(x,y)
```

```python
# Export tree
export_graphviz(dt, out_file='dt.tree', feature_names=['yt','guide'], class_names=['0','1'], filled=True)
```

```python
dt.predict(pd.DataFrame([[1,1]], columns=['yt','guide']))
```

```python
dt.score(x,y)
```
