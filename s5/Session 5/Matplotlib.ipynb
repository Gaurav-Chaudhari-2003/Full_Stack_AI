```python
x = np.random.normal(100, 25, 100)    # (loc, scale, size)
# loc -> mean (center) of the normal distribution (here: 100)
# scale -> standard deviation (spread) of the distribution (here: 25)
# size -> number of random samples to generate (here: 100)


print(x)
plt.hist(x)
plt.show()
```

```python
x = np.array([10,20,30,50])

plt.pie(x, labels = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"], autopct = '%1.1f%%', shadow = True, startangle = 90
        , explode = [0,0,0.2,0.1], colors = ["red","green","blue","yellow"], radius = 1.2,
        wedgeprops = {"edgecolor":"black"}, textprops = {"fontsize":12})
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss,startangle = 90)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.1,0.12,0.13,0.14]
plt.pie(x,labels = labelss,startangle = 90,explode =explodee)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
colorr = ['r','m','gold','g']
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True,colors = colorr)
plt.show() 
```
