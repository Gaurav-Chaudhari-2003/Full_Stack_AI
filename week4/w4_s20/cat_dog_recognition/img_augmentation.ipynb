## Data Augmentation:
* Is the process of generating some subset samples from a one parent source.
* Need: When we have just one image/input/data for a particular classification which is not sufficient to train a model. So we some operaions like: Rotation, Scaling, Transportation, etc. operations to generate the different samples for the same image.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train = ImageDataGenerator(rescale=1./255)
test = ImageDataGenerator(rescale=1./255)

from tensorflow.keras.utils import img_to_array, array_to_img, load_img
data_aug = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.3,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',
    shear_range=0.2,
)


img = load_img("../cat_dog_recognition/assets/actual/cat.png")
img = img_to_array(img)
img = img.reshape((1,) + img.shape)

i = 0
for x in data_aug.flow(img, batch_size=1, save_to_dir="../cat_dog_recognition/assets/augmented/train/cat", save_prefix=f"cat", save_format='png'):
    i+=1
    if i > 30:          # Generate 10 images
        break
```
