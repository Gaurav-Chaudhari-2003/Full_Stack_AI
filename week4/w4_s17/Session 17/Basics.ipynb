```python
data.resample(rule='W').min()       # Week wise data (return week's lowest value)
```

```python
data.resample(rule='ME').std()      # Month End
```

```python
data.resample(rule='MS').count()            # Month Start
```

```python
data.resample(rule='D').max()
```

```python
data.resample(rule='h').mean()      # hour
```

```python
data.resample(rule='MS').max()['Open'].plot()
```

```python
data.resample(rule='YE').max()['Open'].plot(kind = 'bar')        # Year End
```
