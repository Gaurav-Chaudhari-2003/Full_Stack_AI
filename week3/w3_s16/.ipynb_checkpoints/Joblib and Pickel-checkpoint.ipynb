```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data, columns=iris. feature_names)
table.head()
```

```python
table['target'] = iris.target
table.head()
```

```python
table['target'] = iris.target
table. head()
```

```python
X = table.drop(['target'], axis='columns')
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier1 = GaussianNB()
classifier1.fit(X_train, y_train)
```

```python
y_pred=classifier1.predict(X_test)
y_pred
```

```python
classifier1.score(X_test, y_test)
```

```python
import joblib
joblib.dump(classifier1, 'model1')
```

```python
model1 = joblib.load('model1')
y_pred = model1.predict(X_test)
y_pred
```

```python
import pickle
pickle.dump(classifier1, open('model2.pkl','wb'))
```

```python
model2 = pickle.load(open('model2.pkl','rb'))
y_pred = model2.predict(X_test)
y_pred
```
