```python
from sklearn.datasets import make_circles
x, y = make_circles(n_samples=5000,random_state=0,noise=0.03)
plt.scatter(x[:,0],x[:,1],c=y, cmap='viridis')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
clf1 = SVC(kernel="linear")
clf1.fit(x, y)
```

```python
pred = clf1.predict(x)
accuracy_score(pred, y)
```

```python
clf2 = SVC(kernel="poly")
clf2.fit(x, y)
pred = clf2.predict(x)
accuracy_score(pred, y)
```

```python
clf3 = SVC(kernel="rbf")
clf3.fit(x, y)
pred = clf3.predict(x)
accuracy_score(pred, y)
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y, pred)
confusion_matrix
```

```python
cm = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = clf3.classes_)
```

```python
cm.plot()
```

```python
A = np.array([[1,10],[2,8],[3,6],[4,4],[5,6],[6,8],[7,10]])
```

```python
plt.scatter(A[:,0],A[:,1], color='red', s=100, edgecolors='w')
plt.show()
```

```python
x = A[:,0]
y = A[:,1]
```

```python
x = x.reshape(-1,1)
```

```python
from sklearn.svm import SVR 
svr = SVR(kernel='linear') 
svr.fit(x, y)
```

```python
y_pred = svr.predict(x) 
```

```python
plt.scatter(x, y, color='red', marker='o', s=30, label='Original data', edgecolors='white')
plt.plot(x, y_pred, color='green')
svr.score(x,y)
```

```python
svr = SVR(kernel='poly') 
svr.fit(x, y)
y_pred = svr.predict(x) 
plt.scatter(x, y, color='red') 
plt.plot(x, y_pred, color='green')  
svr.score(x,y)
```

```python
svr = SVR(kernel='rbf')
svr.fit(x, y)
y_pred = svr.predict(x) 
plt.scatter(x, y, color='red') 
plt.plot(x, y_pred, color='green')  
svr.score(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly = PolynomialFeatures(degree=5)
poly_x = poly.fit_transform(x.reshape(-1, 1))
model1 = LinearRegression()
model1.fit(poly_x, y)
y_pred = model1.predict(poly_x)
plt.scatter(x, y)
plt.plot(x, y_pred, c='red')
model1.score(poly_x,y)
```
