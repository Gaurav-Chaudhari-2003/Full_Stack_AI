```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import pandas as pd
```

```python
datasett = pd.read_csv('finalreview.csv')
datasett
```

```python
from nltk.corpus import stopwords
import re
corpus = []
for i in range(0, 29):
    review = re.sub('[^a-zA-Z]', ' ', datasett['review'][i])
    review = review.lower()
    review = review.split() 
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]      # only lemmatize the words which are not contained in the English language stopwords library.
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

sentences = []
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        sentences.append(simple_preprocess(sent))

sentences       # get the sentencewise keywords
```

```python
import gensim
# Initialize and train the Word2Vec model
# sentences: The input data (list of tokenized sentences)
# window=3: The maximum distance between the current and predicted word within a sentence
model=gensim.models.Word2Vec(sentences,window=3)        # C-BOW and Skip-Gram

# model.wv.index_to_key contains the list of words in the model's vocabulary
# It is sorted by frequency (most frequent words first)
model.wv.index_to_key
```

```python
from gensim.models import KeyedVectors
word_vectors = model.wv
word_vectors.save("word2vec.wordvectors")

wv = KeyedVectors.load("word2vec.wordvectors")
vector = wv['film']
```
